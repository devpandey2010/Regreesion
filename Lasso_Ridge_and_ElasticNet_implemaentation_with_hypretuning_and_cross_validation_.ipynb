```python
#mpg dataset
#miles per gallon
#1 mile is 1.6km
#1gallon is 3.7litre
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("mpg")
df
#business team eill let you know the target variable
```

```python
df.info()
```

```python
df.drop("name",axis=1,inplace=True)
#why dropping beacuse machine learnning is learning pattern from the data
#and column such as id ,name will not much help to learn pattern from data
#so we drop it
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#since we are not done outlier treatment than better idea would be raplace the missing value with median
df["horsepower"].median()
```

```python
#removed the missing values
df["horsepower"]=df["horsepower"].fillna(df["horsepower"].median())
df["horsepower"]
```

```python
df.info()
```

```python
df.dtypes
```

```python
df["origin"].value_counts()
```

```python
#data encoding
#try to learn this technique
df["origin"]=df["origin"].map({"usa":1,"japan":2,"europe":3})
df["origin"]
```

```python
df['origin'].value_counts()
```

```python
df.info()
```

```python
#separate into x and y
#target variable is mpg
x=df.drop("mpg",axis=1)
y=df["mpg"]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
#scaling optional ask your business team
```

```python
#slr
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#model evaluation
model.coef_
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{model.coef_[i]}")
```

```python
#observations
#coefficient are realtively smaller,if one independent variable changes
#there will be not much difference in prediction
#this is something called as smoother model
#this features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#regularised model
#ridge model
from sklearn.linear_model import Ridge
#in practical implementation lambda is alpha
ridgemodel=Ridge(alpha=1)
ridgemodel
```

```python
ridgemodel.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{ridgemodel.coef_[i]}")
```

```python
#for ridge regression evaluation
y_pred=ridgemodel.predict(x_test)
r2_ridge=r2_score(y_test,y_pred)
r2_ridge
```

```python
#we dont see variation in coefficient of ridge regression as compared to simple linear regression
```

```python
#LASSO REGRESSION
```

```python
from sklearn.linear_model import Lasso
lassomodel=Lasso(alpha=1)
lassomodel
```

```python
lassomodel.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{lassomodel.coef_[i]}")
```

```python
#three features coefficient is 0,lasso helps in feature selection
#model evaluation for lasso
y_predlasso=lassomodel.predict(x_test)
r2_lasso=r2_score(y_test,y_predlasso)
r2_lasso
```

```python
#ELASTIC NET
'''l1_ratio
float, default=0.5
The ElasticNet mixing parameter, with 0 <= l1_ratio <= 1. For l1_ratio = 0
the penalty is an L2 penalty. For l1_ratio = 1 it is an L1 penalty.
For 0 < l1_ratio < 1, the penalty is a combination of L1 and L2.'''
from sklearn.linear_model import ElasticNet
elasticmodel=ElasticNet(alpha=1,l1_ratio=0.5)
elasticmodel

```

```python
elasticmodel.fit(x_train,y_train)
```

```python
#coefficenrt
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{elasticmodel.coef_[i]}")
```

```python
#model evaluation
y_predelastic=elasticmodel.predict(x_test)
r2_elastic=r2_score(y_test,y_predelastic)
r2_elastic
```

```python
#CROSS VALIDATION AND HYPERTUNING
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from  sklearn.linear_model import Lasso,Ridge
```

```python
#define the model
```

```python
lasso=Lasso()
```

```python
#grid search cv>>dictionary with parameters names(str)as keys and lists of parametes settings to try as
'''param_griddict or list of dictionaries
Dictionary with parameters names (str) as keys and lists of parameter settings to try as values,
or a list of such dictionaries, in which case the grids spanned by each dictionary in the list
are explored. This enables searching over any sequence of parameter settings.'''
param_grid={"alpha":[0.001,0.01,0.1,1,10,100]}
param_grid
```

```python
grid_search=GridSearchCV(estimator=lasso,param_grid=param_grid,cv=5,scoring='r2',verbose=2)
```

```python
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
y_pred=grid_search.best_estimator_.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#RANDOM SEARCH CV
random_cv=RandomizedSearchCV(estimator=lasso,param_distributions=param_grid,n_iter=3,cv=5,scoring='neg_root_mean_squared_error',verbose=3)
"""what are the scoring you can find 'roc_auc_ovr', 'positive_likelihood_ratio', 'roc_auc_ovo_weighted', 'jaccard_samples', 'roc_auc_ovo', 'jaccard_micro',
'f1_macro', 'neg_log_loss', 'precision_weighted', 'neg_mean_absolute_error', 'fowlkes_mallows_score', 'neg_brier_score', 'neg_root_mean_squared_log_error',
'precision_macro', 'recall_micro', 'f1', 'mutual_info_score', 'neg_max_error', 'recall_samples', 'homogeneity_score', 'neg_mean_gamma_deviance', 'recall',
 'top_k_accuracy', 'neg_negative_likelihood_ratio', 'accuracy', 'adjusted_rand_score', 'average_precision', 'f1_weighted', 'jaccard_macro', 'normalized_mutual_info_score',
 'r2', 'neg_mean_poisson_deviance', 'f1_samples', 'd2_absolute_error_score', 'jaccard', 'matthews_corrcoef', 'neg_mean_squared_log_error', 'neg_median_absolute_error',
  'neg_mean_absolute_percentage_error', 'f1_micro', 'v_measure_score', 'jaccard_weighted', 'explained_variance', 'recall_macro', 'adjusted_mutual_info_score', 'rand_score',
  'roc_auc_ovr_weighted', 'precision', 'recall_weighted', 'completeness_score', 'roc_auc', 'neg_mean_squared_error', 'precision_micro', 'precision_samples', 'balanced_accuracy', 'neg_root_mean_squared_error'"""
```

```python
random_cv.fit(x_train,y_train)
#due to n_iter parameters in each cross validation random 3 paramets are selected
```

```python
random_cv.best_params_
```

```python
random_cv.best_params_['alpha']
```

```python
random_cv.best_score_
```

```python
random_cv.best_estimator_
```

```python
#using the above alpha or model to predict the on x_test
y_pred=random_cv.best_estimator_.predict(x_test)
neg_mean_squared_error=r2_score(y_test,y_pred)
neg_mean_squared_error
```
