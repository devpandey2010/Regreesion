```python
#mpg dataset
#miles per gallon
#1 mile is 1.6km
#1gallon is 3.7litre
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("mpg")
df
#business team eill let you know the target variable
```

```python
df.info()
```

```python
df.drop("name",axis=1,inplace=True)
#why dropping beacuse machine learnning is learning pattern from the data
#and column such as id ,name will not much help to learn pattern from data
#so we drop it
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#since we are not done outlier treatment than better idea would be raplace the missing value with median
df["horsepower"].median()
```

```python
#removed the missing values
df["horsepower"]=df["horsepower"].fillna(df["horsepower"].median())
df["horsepower"]
```

```python
df.info()
```

```python
df.dtypes
```

```python
df["origin"].value_counts()
```

```python
#data encoding
#try to learn this technique
df["origin"]=df["origin"].map({"usa":1,"japan":2,"europe":3})
df["origin"]
```

```python
df['origin'].value_counts()
```

```python
df.info()
```

```python
#separate into x and y
#target variable is mpg
x=df.drop("mpg",axis=1)
y=df["mpg"]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
#scaling optional ask your business team
```

```python
#slr
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#model evaluation
model.coef_
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{model.coef_[i]}")
```

```python
#observations
#coefficient are realtively smaller,if one independent variable changes
#there will be not much difference in prediction
#this is something called as smoother model
#this features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#regularised model
#ridge model
from sklearn.linear_model import Ridge
#in practical implementation lambda is alpha
ridgemodel=Ridge(alpha=1)
ridgemodel
```

```python
ridgemodel.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{ridgemodel.coef_[i]}")
```

```python
#for ridge regression evaluation
y_pred=ridgemodel.predict(x_test)
r2_ridge=r2_score(y_test,y_pred)
r2_ridge
```

```python
#we dont see variation in coefficient of ridge regression as compared to simple linear regression
```

```python
#LASSO REGRESSION
```

```python
from sklearn.linear_model import Lasso
lassomodel=Lasso(alpha=1)
lassomodel
```

```python
lassomodel.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{lassomodel.coef_[i]}")
```

```python
#three features coefficient is 0,lasso helps in feature selection
#model evaluation for lasso
y_predlasso=lassomodel.predict(x_test)
r2_lasso=r2_score(y_test,y_predlasso)
r2_lasso
```

```python
#ELASTIC NET
'''l1_ratio
float, default=0.5
The ElasticNet mixing parameter, with 0 <= l1_ratio <= 1. For l1_ratio = 0
the penalty is an L2 penalty. For l1_ratio = 1 it is an L1 penalty.
For 0 < l1_ratio < 1, the penalty is a combination of L1 and L2.'''
from sklearn.linear_model import ElasticNet
elasticmodel=ElasticNet(alpha=1,l1_ratio=0.5)
elasticmodel

```

```python
elasticmodel.fit(x_train,y_train)
```

```python
#coefficenrt
for i,col_name in enumerate(x_train.columns):
  print(f"{col_name}:{elasticmodel.coef_[i]}")
```

```python
#model evaluation
y_predelastic=elasticmodel.predict(x_test)
r2_elastic=r2_score(y_test,y_predelastic)
r2_elastic
```
