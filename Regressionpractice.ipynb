```python
"""Q1.Write a Python script to visualize the distribution of errors (residuals) for a multiple linear regression model
using Seaborn's "diamonds" dataset"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

data=sns.load_dataset('diamonds')
data
features=['carat','depth','table','x','y','z']
target='price'
#split the data in x and y
x=data[features]
y=data[target]
# split the data into train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#scaling can be done but it is optional
#model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#prediction
y_pred=lr.predict(x_test)
#calculate error
error=y_test-y_pred
#plot the error
plt.figure(figsize=(10,6))
sns.histplot(error,kde=True)
plt.title('Distribution of Errors')
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.show()
```

```python
'''Q2Write a Python script to calculate and print Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root
Mean Squared Error (RMSE) for a linear regression model.'''
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
df=pd.DataFrame(data.data,columns=data.feature_names)
df["target"]=data.target
#multiple regression
#dividing into x and y
x=df.drop("target",axis=1)
y=df["target"]
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
#model selection
from sklearn.linear_model  import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
#model evaluate
y_pred=model.predict(x_test)
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("Mean Squared Error:",mse)
print("Mean Absolute Error:",mae)
print("Root Mean Squared Error:",rmse)
```

```python
"""Q3.Write a Python script to check if the assumptions of linear regression are met. Use a scatter plot to check
linearity, residuals plot for homoscedasticity, and correlation matrix for multicollinearity"""
data=sns.load_dataset('diamonds')
data
features=['carat','depth','table','x','y','z']
target='price'
#split the data in x and y
x=data[features]
y=data[target]
# split the data into train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#scaling can be done but it is optional
#model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#prediction
y_pred=lr.predict(x_test)
#linearity
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Linearity")
plt.show()
#residual plot
error=y_test-y_pred
plt.scatter(y_pred,error)
plt.axhline(y=0,color='r',linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.show()
#Multicollinearity
corr_matrix=x_train.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
'''5Implement a simple linear regression model on a dataset and print the model's coefficients, intercept, and
R-squared score.'''

from sklearn.datasets import load_diabetes
data=load_diabetes()
df=pd.DataFrame(data.data,columns=data.feature_names)
df["target"]=data.target
#i want to check how bp will predict the target
x=df[["bp"]]
y=df["target"]
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#prediction
y_pred=lr.predict(x_test)
#coefficient
print("Coefficient:",lr.coef_)
#intercept
print("Intercept:",lr.intercept_)
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("R-squared score:",r2)
```

```python
'''6. Write a Python script that analyzes the relationship between total bill and tip in the 'tips' dataset using
simple linear regression and visualizes the results.'''
data=sns.load_dataset("tips")
data
df=pd.DataFrame(data)
#simple linear regression
x=df[["total_bill"]]
y=df["tip"]
x
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#model evaluation
y_pred=lr.predict(x_test)
error=y_test-y_pred
#plot the error
plt.figure(figsize=(10,6))
sns.histplot(error,kde=True)
plt.title('Distribution of Errors')
plt.xlabel('Error')
plt.ylabel('Frequency')
#residual
plt.scatter(y_pred,error)
plt.axhline(y=0,color='r',linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.show()
```

```python
'''7. Write a Python script that fits a linear regression model to a synthetic dataset with one feature. Use the
model to predict new values and plot the data points along with the regression line'''
np.random.seed(42)
x=np.random.rand(100,1)
y=4+3*x+np.random.rand(100,1) # linear relationship y=mx+c+error/noise
#split data into train  test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#model training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
#model evaluation
y_pred=model.predict(x)
#plot
plt.scatter(x,y,color='blue',label='Actual')
plt.plot(x,y_pred,color='red',label='Predicted')
plt.xlabel("featutre")
plt.ylabel("Target")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
'''9. Write a Python script that fits a polynomial regression model (degree 2) to a dataset and plots the
regression curve.'''
data=sns.load_dataset("diamonds")
data
features=['x']  #we can plot by changing x
target='price'
#split the data in x and y
x=data[features]
y=data[target]
#model training
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
#generate predictions
x_range = np.linspace(x.min(), x.max(), 500).reshape(-1, 1)
x_range_poly = poly.transform(x_range)
y_pred = model.predict(x_range_poly)
plt.scatter(x,y,color='blue',label='Actual')
plt.plot(x_range,y_pred,color='red', label="Degree-2 Polynomial Fit")
```
